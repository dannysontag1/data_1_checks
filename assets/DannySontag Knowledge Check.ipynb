```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("pokemon_data.csv")
df.head(5)
```

```python
df.columns
```

```python
attack = df['Attack']
defense = df['Defense']
```

```python
sns.scatterplot(x=attack, y=defense)
```

```python
df_Type1_group = df.groupby(['Type 1'], as_index = False).sum()
df_Type1_group
```

```python
plt.figure(figsize = (18,6))
plt.bar(df_Type1_group['Type 1'], df_Type1_group['Legendary'])
plt.xlabel('Type 1')
plt.ylabel('Lengendary')
plt.show()
```
